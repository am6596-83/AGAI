# Question 1:
Events A and B are independent, with P(A) = 0.4 and P(B) = 0.3. Calculate:

**(a)** P(A ∩ B)

**(b)** P(A ∪ B)

```python
# Given
P_A = 0.4
P_B = 0.3

# (a) P(A ∩ B) - Since A and B are independent
P_A_and_B = P_A * P_B
print(f"(a) P(A ∩ B) = {P_A} × {P_B} = {P_A_and_B}")

# (b) P(A ∪ B) - Addition Rule
P_A_or_B = P_A + P_B - P_A_and_B
print(f"(b) P(A ∪ B) = {P_A} + {P_B} - {P_A_and_B} = {P_A_or_B}")
```

## Question 2
**Suppose P(A) = 0.5, P(B) = 0.4, and P(A|B) = 0.7. Are events A and B independent?**

To test for independence, we use the probability rule for independence:

**Independence Rule:** P(A ∩ B) = P(A) × P(B) if A and B are independent

First, we calculate P(A ∩ B) assuming independence, then we use the 
Conditional Probability rule to calculate what P(A|B) would be under that assumption:

**Conditional Probability:** P(A|B) = P(A ∩ B) / P(B)

Because the calculated P(A|B) does not match the given P(A|B) = 0.7, the events are not independent.
See calculation below.

```python
# Question 2 Code
# Suppose / Given
P_A = 0.5
P_B = 0.4
P_A_given_B = 0.7

# Independence: P(A ∩ B) = P(A) × P(B) if A and B are independent
P_A_and_B = P_A * P_B
print(f"P(A ∩ B) = {P_A} × {P_B} = {P_A_and_B}")

# Conditional Probability: P(A|B) = P (A∩B) / P(B)
P_A_given_B_calculated = P_A_and_B / P_B
print(f"P(A|B) given = {P_A_given_B}")
print(f"P(A|B) calculated = P(A ∩ B) / P(B) = {P_A_and_B} / {P_B} = {P_A_given_B_calculated}")
print(f"Since {P_A_given_B} ≠ {P_A_given_B_calculated}, A and B are NOT independent")
```

**Answer 2: No, A and B are not independent**

# Question 3: 
Use Bayes’ Rule to solve the following: Given P(A) = 0.6, P(B|A) = 0.5, and P(B|A′) = 0.2, calculate P(A|B).

```python
# Given
P_A = 0.6
P_B_given_A = 0.5
P_B_given_A_complement = 0.2

# Law of Total Probability: P(B) = P(B|A) × P(A) + P(B|A') × P(A')
P_A_complement = 1 - P_A
P_B = P_B_given_A * P_A + P_B_given_A_complement * P_A_complement
print(f"P(A') = 1 - {P_A} = {P_A_complement}")
print(f"P(B) = {P_B_given_A} × {P_A} + {P_B_given_A_complement} × {P_A_complement} = {P_B}")

# Bayes' Rule: P(A|B) = P(B|A) × P(A) / P(B)
P_A_given_B = P_B_given_A * P_A / P_B
print(f"P(A|B) = P(B|A) × P(A) / P(B) = {P_B_given_A} × {P_A} / {P_B} = {P_A_given_B:.4f}")
```

# Question 4:
A medical test for a disease is 95% accurate for detecting the disease when a
person actually has it (true positive rate), and it correctly gives a negative result
for 90% of the healthy people (true negative rate). If the probability of having
the disease is 2%, calculate the probability that a person who tests positive
actually has the disease. (Use Bayes’ Rule).

```python
# Given
P_disease = 0.02          # P(A): probability of having the disease
P_positive_given_disease = 0.95   # P(B|A): true positive rate
P_negative_given_healthy = 0.90   # true negative rate
P_healthy = 1 - P_disease         # P(A')
P_positive_given_healthy = 1 - P_negative_given_healthy   # P(B|A'): probability of testing positive given healthy (false positive rate)

# Law of Total Probability: P(B) = P(B|A) × P(A) + P(B|A') × P(A')
P_positive = P_positive_given_disease * P_disease + P_positive_given_healthy * P_healthy
print(f"P(A') = 1 - {P_disease} = {P_healthy}")
print(f"P(B|A') = 1 - {P_negative_given_healthy} = {P_positive_given_healthy:.4f}")
print(f"P(B) = {P_positive_given_disease} × {P_disease} + {P_positive_given_healthy:.4f} × {P_healthy} = {P_positive:.4f}")

# Bayes' Rule: P(A|B) = P(B|A) × P(A) / P(B)
P_disease_given_positive = P_positive_given_disease * P_disease / P_positive
print(f"Probability that a person who tests positive actually has the disease (to 4 decimal points): P(A|B) = {P_positive_given_disease} × {P_disease} / {P_positive:.4f} = {P_disease_given_positive:.4f}")

#Cleaned up with rounding to 4 decimal places.
```

# Question 5
A discrete random variable X represents exam scores with the following probability distribution:

| X | 85 | 90 | 95 | 100 |
|---|----|----|----|----|
| P(X=x) | 0.375 | 0.375 | 0.125 | 0.125 |

**(a)** Calculate the expected value $E[X]$

**(b)** Calculate the variance $Var(X) = E[X^2] - (E[X])^2$

**(c)** The following sample was drawn from this distribution: {85, 90, 85, 95, 90, 85, 100, 90}. Calculate the sample mean and compare it with the expected value from part (a).

```python
# Given
scores = [85, 90, 95, 100]
probs = [0.375, 0.375, 0.125, 0.125]

# (a) Expected Value: E[X] = Σ x · P(X = x)
E_X = sum(x * p for x, p in zip(scores, probs))
print(f"(a) E[X] = {scores[0]}×{probs[0]} + {scores[1]}×{probs[1]} + {scores[2]}×{probs[2]} + {scores[3]}×{probs[3]} = {E_X:.2f}")

# (b) Variance: Var(X) = E[X²] - (E[X])²
E_X2 = sum(x**2 * p for x, p in zip(scores, probs))
Var_X = E_X2 - E_X**2
print(f"(b) E[X²] = {E_X2:.2f}")
print(f"    Var(X) = E[X²] - (E[X])² = {E_X2:.2f} - {E_X:.2f}² = {Var_X:.2f}")

# (c) Sample mean
sample = [85, 90, 85, 95, 90, 85, 100, 90]
sample_mean = sum(sample) / len(sample)
print(f"(c) Sample mean = {sum(sample)} / {len(sample)} = {sample_mean:.2f}")
print(f"    Expected value E[X] = {E_X:.2f}")
print(f"    Difference = {abs(sample_mean - E_X):.2f}")
```

## Question 6
A communication system transmits one of four possible messages with the following probabilities:

| Message | A | B | C | D |
|---------|-----|-----|-----|-----|
| Probability | 0.4 | 0.3 | 0.2 | 0.1 |

**(a)** Calculate the entropy $H(X)$ of this message distribution in bits (use $log_2$)

$H(X) = -\sum P(X=x) \log_2 P(X=x)$

**(b)** If we could redesign the system to have equal probabilities for all messages, what would be the new entropy?

```python
import math

# Given
probs = [0.4, 0.3, 0.2, 0.1]
messages = ['A', 'B', 'C', 'D']

# (a) Entropy: H(X) = -Σ P(X=x) log2(P(X=x))
H_X = -sum(p * math.log2(p) for p in probs)
print(f"(a) H(X) = -{' + '.join([f'{p}×log2({p})' for p in probs])}")
print(f"    H(X) = {H_X:.2f} bits")

# (b) Equal probabilities for 4 messages
p_equal = 1/4
H_equal = -sum(p_equal * math.log2(p_equal) for _ in probs)
print(f"\n(b) Equal probability for each message = 1/4 = {p_equal}")
print(f"    H(X) = -{4}×{p_equal}×log2({p_equal}) = {H_equal:.2f} bits")
```
